```python
def replace_missing_value_with_NASA_data(dataframe, nasaData):
    """This function replaces missing value from NIMET data with the corresponding value in NASA data.
    First, it prepares the NASA date for conform with the format of NIMET data,
    It then fills mssing values from NIMET data column-wise
    It joins the updated data to the NIMET data"""
    # Preparing NASA data to conform with NIMET data
    needed_data = nasaData[['YEAR', 'MO','DY','T2M_MIN','T2M_MAX','PRECTOT']].copy()
    needed_data.rename(columns= {'YEAR':'Year','MO':'month','DY':'Day','T2M_MIN':'MinTemp','T2M_MAX':'MaxTemp','PRECTOT':'Precipitn'}, inplace = True)
    needed_data['Date'] = pd.to_datetime(needed_data[['Year', 'month', 'Day']])
    needed_data = needed_data[['Date','MaxTemp','MinTemp','Precipitn','month','Day','Year']]
    needed_data.index=list(range(1608, 15761))
    
    # 
    ret_dataframe = dataframe.copy()
    new_ikeja = dataframe.copy()
    updated_precip = new_ikeja.set_index("Date").Precipitn.fillna(needed_data.set_index("Date").Precipitn).reset_index()
    updated_maxtemp =  new_ikeja.set_index("Date").MaxTemp.fillna(needed_data.set_index("Date").MaxTemp).reset_index()
    updated_mintemp =  new_ikeja.set_index("Date").MinTemp.fillna(needed_data.set_index("Date").MinTemp).reset_index()
    
    ret_dataframe['Precipitn']= updated_precip['Precipitn']
    ret_dataframe['MaxTemp']= updated_maxtemp['MaxTemp']
    ret_dataframe['MinTemp']= updated_mintemp['MinTemp']
    
    return ret_dataframe
```

```python
pre_clean_data = replace_missing_value_with_NASA_data(ikeja_daily, Nasa_ikeja)
```

```python
# Observations
# Remaining missing values are the whole month of December 1980 and a day in May 1978
# We can fill that with the mode of the data grouped by month and day
```

```python
def replace_missing_with_mode(dataframe):
    df_copy = dataframe.copy()
    missing_month= df_copy[(df_copy.Precipitn.isnull()) | (df_copy.MaxTemp.isnull())|(df_copy.MinTemp.isnull()) ]['month'].tolist()
    missing_day = df_copy[(df_copy.Precipitn.isnull()) | (df_copy.MaxTemp.isnull())|(df_copy.MinTemp.isnull())]['Day'].tolist()
    missing_dates = zip(missing_month, missing_day)
    for i in missing_dates:
        replacement = df_copy[(df_copy.month==i[0]) & (df_copy.Day==i[1])].mode().iloc[0]
        df_copy[(df_copy.month==i[0]) & (df_copy.Day==i[1])] = df_copy[(df_copy.month==i[0]) & (df_copy.Day==i[1])].fillna(replacement)
    print('Succesful replacement')    
    return df_copy
```

```python
# Check the seasonality and how stationary the data is
def check_stationary(data):
    from statsmodels.tsa.seasonal import seasonal_decompose
    stat_data = data.copy()
    stat_data = stat_data.set_index('Date')
    result = seasonal_decompose(stat_data['Precipitn'], model='additive', freq= 365)
    fig = plt.figure()
    fig = result.plot()
    fig.set_size_inches(15,12)
```

```python
check_stationary(clean_data)
```

```python
# Yearly pattern is obvious from above plot
# The mean and variance of the data should not be a function of time. Meaning the data is not stationary
# To confirm that the data is stationary or not, we can use the Dickey-Fuller test.
# The assumption of Dickey fuller is that if the test statistic is greater than the critical value, then the time series is stationary
```

```python
def test_stationarity(timeseries):

    #Determine rolling statistics
    # rolmean = pd.rolling_mean(timeseries, window = 12)
    rolmean = timeseries.rolling(window = 12).mean()
    # rolstd = pd.rolling_std(timeseries, window = 12)
    rolstd = timeseries.rolling(window = 12).std()

    #plot rolling data
    plt.figure(figsize=(15,4))
    orig = plt.plot(timeseries, color = 'blue', label = 'Original')
    mean = plt.plot(rolmean, color = 'red', label = 'Rolling Mean')
    std = plt.plot(rolstd, color = 'black', label = 'Rolling std')
    plt.title("Rolling Mean and Standard Deviation")
    plt.legend(loc = 'best')

    #Dickey - Fuller test
    dftest = adfuller(timeseries, autolag = 'AIC')
    dfoutput = pd.Series(dftest[0:4], index = ['Test Statistic', 'p-value', 'Lags Used', 'Number of observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value(%s)'%key] = value
    print (dfoutput)
```

```python
test_stationarity(clean_data['Precipitn'])
```

```python
def evaluate_arima_model(X, arima_order):
    """This function runs ARIMA model on the a dataset. It splits the data into training and test,
    fits the model and predict one value in the future."""
    train_len = int(len(X)*0.66)
    train, test = X[0:train_len], X[train_len:]
    predictions = list()
    for t in range(len(test)):
        history = X[0: train_len]
        model = ARIMA(history, order = arima_order)
        model_fit = model.fit(disp = -1)
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        train_len = train_len+1
    error = mean_squared_error(test, predictions)
    return error
```

```python
def evaluate_models(datasets, p_d_q_values):
    best_score, best_cfg = float("inf"), None
    for i in p_d_q_values:
        order = (i[0],i[1],i[2])
        print('trying order', order)
        try:
            mse = evaluate_arima_model(datasets, order)
            if mse<best_score:
                best_score, best_cfg = mse, order
                print("ARIMA%s MSE:%.3f" % (order, mse))
        except:
            continue
    print("Best ARIMA%s MSE:%.3f" % (best_cfg, best_score))
```

```python
arima_data = clean_data[['Date', 'Precipitn']]
arima_data.set_index('Date', inplace=True)
```
